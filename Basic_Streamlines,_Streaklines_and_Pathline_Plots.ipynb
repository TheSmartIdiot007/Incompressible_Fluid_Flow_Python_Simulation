```python
times = np.linspace(0,5,100)
```

```python
x= np.linspace(-2,2,20)
y =np.linspace(-2,2,20)
def vel(r,t):
    x,y=r
    if((x**2 + y**2) < 1):
        return 0,0
    u = ((x**2 - y**2 -1)*(x**2 - y**2)+ 4*(x**2)*(y**2))/(x**2 + y**2)**2
    v = -2*y*x/(x**2 + y**2)**2
    if((x**2 + y**2) >= 1):
        return u,v
```

```python

n = np.linspace(-2,2,10)
for i in range(10):
    ic = np.array([-2, n[i]])
    a, b = odeint(vel, ic, times).T
    plt.plot(a,b, '-b')
    plt.xlabel('x')
    plt.ylabel('y')
#Pathlines 
```

```python
for i in range(10):
    ic = np.array([-2,-2+0.4*i ])
    dt = 0.1
    n = 100
    T = n*dt
    sx, sy = [], []
    for j in range(n):
        start = j*dt
        res = odeint(vel, ic, [start, T])
        sx.append(res[1,0])
        sy.append(res[1,1])
    plt.plot(sx, sy, '-b')
    plt.grid()
    plt.axis('equal');
#Streaklines
```
