```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
from math import *
```

```python
def v_vortex(g, r):
  x,y = r
  return [-(g*y)/(2*pi*(x**2 + y**2)),(g*x)/(2*pi*(x**2 + y**2))]
```

Q1. Compute the coefficient of pressure, Cp along the surface of the cylinder as a function of θ from the x-axis. Plot this coefficient of pressure. Also plot the surface velocity magnitude distribution as a function of θ.

Answer: For this we already have solved for the value of $Φ$ in the class, so I am directly using it.
$$ \Phi  = U_\infty z + {{U_\infty R^2 } \over z}$$
From this we get that $$u =  U_\infty + {{U_\infty R^2 (y^2 - x^2)} \over (x^2 + y^2)^2} $$
&
$$v =-{{2 U_\infty R^2 xy} \over (x^2 + y^2)^2} $$

```python
def v_cylinder(u_0,R,r):
  x, y = r
  return [u_0 - (u_0*R**2)*(x**2 - y**2)/(x**2 + y**2)**2, -(u_0*R**2)*(2*x*y)/(x**2 + y**2)**2]
```

```python
R = 1
x,y= np.mgrid[-2:2:20j, -2:2:20j]
r = [x,y]
u = np.ones_like(x)
v = np.ones_like(y)
cd = (x**2 + y**2) < R**2
u_0 = 1
u,v = v_cylinder(u_0, R, r)
u[cd] = 0
v[cd]= 0
plt.clf()
plt.quiver(x, y, u, v);
```

From Bernoulli we get that:
$$ p_\infty + {1 \over 2} \rho U_\infty ^2 = p_s + {1 \over 2} \rho
 v_s^2$$
 & as we know that 
 $$ C_p = {{p_s - p_∞} \over {{1} \over 2} ρ U_\infty ^2} $$
 Both of above equations gives us that:
 $$ C_p = {{x^2 - 3y^2} \over x^2 + y^2 } = 1 - 4 \cos ^2 θ$$

```python
theta = np.linspace(0,2*pi, 1000)
c_p = 1-4*(np.cos(theta))**2
plt.plot(theta, c_p)
```

```python
#Velocity magnitude distribution:
rd = (x**2 + y**2)**(0.5)
cos = x/rd
sin = y/rd
#As we know that the radial velocity will be zero at the surface which means that we have to only concentrate on the angular velocity component
v_s = v*cos - u*sin
#If we convert it into polar coordinates then we get that:
theta = np.linspace(0,2*pi,1000)
v_s = -2*u_0*np.sin(theta) 
plt.plot(theta,v_s)
```

Q4. Plot the pressure field all over the domain.

From Bernoulli we get that:
$$ p_\infty + {1 \over 2} \rho U_\infty ^2 = p + {1 \over 2} \rho
 v_s^2$$
Which is same as:
$$p = p_\infty + {1 \over 2} \rho U_\infty ^2 - {1 \over 2} \rho
 v^2 $$

```python
#From above result we get that:
x,y= np.mgrid[-6:6:1000j, -6:6:1000j]
p_0 = 101325 #pressure at infinity
g = 100000 #vortex strength
u_0 = 330 #speed at infinity
rho = 1 #density
R = 1
r = [x,y]
u1 = np.ones_like(x)
v1 = np.ones_like(y)
u2 = np.ones_like(x)
v2 = np.ones_like(y)
u = np.ones_like(x)
v = np.ones_like(y)
cd = (x**2 + y**2) < R**2
g=100
u1, v1= v_cylinder(u_0, R, r)
u2,v2 = v_vortex(g, r)
u = u1 + u2
v = v1+ v2
p = p_0 + 0.5*rho*u_0**2 - 0.5*rho*(u**2 + v**2)
p[cd] = 0
plt.contourf(x, y, p, 20, cmap='RdGy')
plt.colorbar();
```

Q7. Integrate this Cp distribution as a function of theta by hand and use the resulting expression to find the lift and drag forces on the cylinder. You can use a simple mid-point rule to perform this quadrature, i.e. you can numerically compute the integral by dividing the interval 0,2π into equal parts, and finding the integral as the sum of the value of the function at the mid-point of each interval into the interval length.


I would like to break this in two part:

i. When vortex strength is low (Γ < 1000): It converges to 2π (Independednt of vortex strength)



```python
#Integrating c_p
g = 1000 #vortex strength
R=1
p_0 = 101325 #pressure at infinity
u_0 = 330 #speed at infinity
rho = 1 #density
n=10000000
theta = np.linspace(0,2*pi,n)
C_p_surface =  1-(1/u_0**2)*(g/(4*pi**2*R**2) + 2*u_0**2 -2*np.cos(2*theta)*u_0**2 -2*g*np.sin(theta)*u_0/(pi*R))
I_c_p = sum(C_p_surface)*(2*pi/n)
print(I_c_p)
```

ii. When vortex strength is high : It converges to some other value depending on vortex strength

```python
#Integrating c_p
g = 1000000 #vortex strength
R=1
p_0 = 101325 #pressure at infinity
u_0 = 330 #speed at infinity
rho = 1 #density
n=10000000
theta = np.linspace(0,2*pi,n)
C_p_surface =  1-(1/u_0**2)*(g/(4*pi**2*R**2) + 2*u_0**2 -2*np.cos(2*theta)*u_0**2 -2*g*np.sin(theta)*u_0/(pi*R))
I_c_p = sum(C_p_surface)*(2*pi/n)
print(I_c_p)
```
